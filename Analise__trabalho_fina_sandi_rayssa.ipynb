```python
import pandas as pd
import matplotlib.pyplot as plt
df_estado = pd.read_csv("sandi_rayssa_estados_ferro.csv")
df_estado.columns = ['UF', 'Total']

uf = df_estado['UF']
total = df_estado['Total']
fig, ax = plt.subplots(figsize =(12, 8))
plt.bar(uf, total, color="blue")

ax.set_title("Estados possuem Mineração de Ferro - Barragens") 
plt.show()

```

```python
# produção 

df_producao = pd.read_csv("sandi_rayssa_beneficiada_analise_producao.csv")
df_producao['Produção'] = df_producao['Produção'].str.replace(',', '.')
df_producao['Produção'] = df_producao['Produção'].astype('float')
pd.options.display.float_format = '{:.2f}'.format 
df_producao
```

```python
# Definindo variáveis
y = df_producao['Produção']
x = df_producao['Ano']

plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)

 
# Criando um gráfico
plt.bar(x, y, color="blue")
#plt.xticks(rotation=45)

# nomes = df_producao.index.tolist()
# plt.xticks(range(len(nomes)), nomes, rotation = 0)


# Atribuindo um título ao gráfico
plt.title('Produção em tonelada de Minério de Ferro por ano')
plt.xlabel('Anos')
plt.ylabel('Milhões de Toneladas')
```

```python
## Análise vendas por ano - Minério de ferro

df_ano_venda = pd.read_csv("sandi_rayssa_beneficiada_analise_venda_2.csv")
df_ano_venda
```

```python
ano = df_ano_venda['Ano']
valor = df_ano_venda['Valor Vendido em Reais']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("Valor total vendido")
plt.xlabel("Anos")
plt.title("Valor de Vendas do estado de MG")
plt.bar(ano, valor, color="blue")
plt.show()
```

```python
## Análise quantidade vendida por ano - Minério de ferro

df_ano_quantidade = pd.read_csv("sandi_rayssa_beneficiada_analise_venda_1.csv")
df_ano_quantidade['Quantidade de Vendas'] = df_ano_quantidade['Quantidade de Vendas'].str.replace(',', '.')
df_ano_quantidade['Quantidade de Vendas'] = df_ano_quantidade['Quantidade de Vendas'].astype('float')
pd.options.display.float_format = '{:.2f}'.format 
df_ano_quantidade
```

```python
ano = df_ano_quantidade['Ano']
valor = df_ano_quantidade['Quantidade de Vendas']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("Quantidade total vendida")
plt.xlabel("Anos")
plt.title("Quantidade de vendas em toneladas do estado de MG")
plt.bar(ano, valor, color="blue")
plt.show()
```
