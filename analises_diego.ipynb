```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_autuacao_1 = pd.read_csv("diego_autuacao_analise_1.csv")

df_autuacao_1['nome_titular']= df_autuacao_1['nome_titular'].replace(['PORTO UNIAO EXTRACAO DE AREIA LTDA', 'Cbe Companhia Brasileira de Equipamento', 'EMPRESA DE MINERACAO PANORAMA LTDA', 'Coagro Comércio de Areia Grossa Ltda. Epp', 'Porto de Areia Ilha Carolina Ltda Epp'],['PORTO UNIAO','cbe', 'PANORAMA', 'Coagro', 'Porto A Ilha Carolina'])
df_autuacao_1['nome_titular'] = df_autuacao_1['nome_titular'].str.lower()
df_autuacao_1['nome_titular']= df_autuacao_1['nome_titular'].replace(['mineração caldense ltda', 'cubatão dragagens ltda.', 'votorantim cimentos brasil s a', 'mineração curimbaba ltda' ],['Caldense', 'cubatão dragagens', 'votorantim', 'curimbaba'])


df_autuacao_1
```

```python
empresa = df_autuacao_1['nome_titular']
quantidade = df_autuacao_1['qtd']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("quantidade")
plt.xlabel("empresa")
plt.title("Autuações realizadas por empresa")
plt.bar(empresa, quantidade, color="#adb5bd")
plt.xticks(rotation=75)
plt.show()
```

```python
#valor_distribuicao_ferro_2020
df_distrib_1 = pd.read_csv("diego_distribuicao_analise_4.csv")
df_distrib_1
```

```python

df_distrib_1['valor'] = df_distrib_1['valor'].astype('float')
pd.options.display.float_format = '{:.2f}'.format 

df_distrib_1
```

```python
mes = df_distrib_1['mes']
valor = df_distrib_1['valor']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("valor")
plt.xlabel("mês")
plt.title("Distribuição de valores do Ferro em 2020")
plt.bar(mes, valor, color="#adb5bd")
# plt.xticks(rotation=75)
plt.show()
```
