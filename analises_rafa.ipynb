```python
ano =df_rafa_7['ano_base']
valor = df_rafa_7['valor_venda_rs']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("valor")
plt.xlabel("ano")
plt.title("Valor vendido de carvão no RS entre 2010 e 2020")
plt.bar(ano, valor, color="#adb5bd")
# plt.xticks(rotation=75)
plt.show()
```

```python
df_rafa_8 = pd.read_csv("7_analise_rafa.csv")
df_rafa_8.drop('Unnamed: 0', inplace=True, axis=1)

df_rafa_8['valor_venda_rs'] = df_rafa_8['valor_venda_rs'].astype('float')
pd.options.display.float_format = '{:.2f}'.format 

df_rafa_8 = df_rafa_8.groupby(['ano_base'])['valor_venda_rs'].sum().reset_index()

df_rafa_8 = pd.DataFrame(df_rafa_8)

df_rafa_8
```

```python
ano =df_rafa_8['ano_base']
valor = df_rafa_8['valor_venda_rs']
plt.figure(figsize=(10, 5))
plt.gcf().set_size_inches(12, 8)
plt.ylabel("valor")
plt.xlabel("ano")
plt.title("Valor vendido de ferro no Brasil entre 2010 e 2020")
plt.bar(ano, valor, color="#adb5bd")
# plt.xticks(rotation=75)
plt.show()
```
